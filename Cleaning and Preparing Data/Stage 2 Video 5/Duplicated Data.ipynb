```python
import pandas as pd
import numpy as np
```

```python
demo = pd.read_csv('Demographics.csv')
bmx = pd.read_csv('BodyMeasures.csv')
ocq = pd.read_csv('Occupation.csv')
```

```python
demo.sort_values('SEQN',inplace=True)
```

```python
ind = demo['SEQN'].duplicated(keep=False)
```

```python
sum(ind)
```

```python
print('Duplicated rows : %i'% len(demo.loc[ind,:]))
demo.loc[ind,:].head()
```

```python
#drop duplicates only drops rows where every value is exactly the same
demo.drop_duplicates(inplace=True)
```

```python
#there are still more duplicates we have to handle
ind = demo['SEQN'].duplicated(keep=False)
print('Duplicated rows : %i'% len(demo.loc[ind,:]))
demo.loc[ind,:].head()
```

```python
#now we want to handle duplicate rows where the values are spread out over two rows
import itertools
```

```python
dup_id = demo.loc[ind,'SEQN'].unique()
```

```python
for id in dup_id:
    dup_rows=np.where(demo['SEQN']==id)[0]
    for(row1,row2) in itertools.product(dup_rows,repeat=2):
        demo.iloc[row1,:].fillna(demo.iloc[row2,:],axis=0)
```

```python
demo.drop_duplicates(inplace=True)
ind=demo['SEQN'].duplicated(keep=False)
print('Duplicate rows: %i ' % len(demo.loc[ind,:]))
demo.loc[ind,:].head()
```

```python
# remaining duplicates have conflicting values so it is safest to drop all remaining rows
ind= demo['SEQN'].duplicated(keep=False)
demo=demo.loc[~ind,:]
```
